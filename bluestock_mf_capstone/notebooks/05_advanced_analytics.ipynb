```
# Bonus 3: Monte Carlo Simulation
def monte_carlo_simulation(returns, days=252*5, sims=100):
    mean_ret = returns.mean()
    std_ret = returns.std()
    simulated_paths = np.zeros((days, sims))
    simulated_paths[0] = 1 # Start at normalized NAV 1
    for t in range(1, days):
        rand_shocks = np.random.normal(mean_ret, std_ret, sims)
        simulated_paths[t] = simulated_paths[t-1] * (1 + rand_shocks)
    return simulated_paths
```
