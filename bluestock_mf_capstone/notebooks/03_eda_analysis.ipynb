# Exploratory Data Analysis (EDA)
In this notebook, we visualize data trends, SIP inflows, and AUM by fund house.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

processed_dir = Path('../data/processed')
```

```
aum = pd.read_csv(processed_dir / '03_aum_by_fund_house.csv')
sns.barplot(data=aum.head(10), x='aum_crores', y='fund_house')
plt.title('Top 10 Fund Houses by AUM')
plt.show()
```
