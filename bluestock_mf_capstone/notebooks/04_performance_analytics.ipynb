# Performance Analytics
Here we compute CAGR, Sharpe Ratio, Beta, and Value at Risk (VaR). 

*Important:* We annualize using 252 trading days.

```
import pandas as pd
import numpy as np
from pathlib import Path

processed_dir = Path('../data/processed')
nav = pd.read_csv(processed_dir / '02_nav_history.csv')
nav['date'] = pd.to_datetime(nav['date'])
```

```
def calculate_cagr(start_val, end_val, periods, freq='D'):
    years = periods / 252 if freq == 'D' else periods
    if start_val == 0:
        return np.nan
    return (end_val / start_val)**(1/years) - 1
```

```
def calculate_sharpe(returns, risk_free_rate=0.07):
    excess_return = returns.mean() * 252 - risk_free_rate
    volatility = returns.std() * np.sqrt(252)
    if volatility == 0:
        return 0
    return excess_return / volatility
```
